```python
import random

def get_5_students(student_list):
    
    selected_students = random.sample(student_list, 6)
    
    # Remove selected students from the original list
    student_list = [student for student in student_list if student not in selected_students]
    
    # Return the selected students and updated list
    return selected_students, student_list
```
